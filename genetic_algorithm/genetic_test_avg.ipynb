```python
from sudoku import Sudoku

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time

PATH = './sudoku_hard.txt'
CANDIDATES_NR = 100
GENERATIONS = 1000
MUTATION_RATE = 0.6
SELECTION_RATE = 0.85
ELITE_SIZE = 10
```

```python
def solve(cand_nr=CANDIDATES_NR, mutation_rate=MUTATION_RATE, elite_size=ELITE_SIZE):
    sudoku = Sudoku(PATH, cand_nr, GENERATIONS, mutation_rate, SELECTION_RATE, elite_size)
    return sudoku.solve()
```

## Number of candidates

```python
candidates = [10, 20, 30, 50, 75, 100, 150, 200, 300, 500]
results = list()
times = list()

for cand in candidates:
    temp = list()
    start_time = time.time()
    for _ in range(10):
        temp.append(solve(cand_nr=cand, elite_size=int(cand*0.2)))
    results.append(np.mean(temp))
    end_time = time.time()
    times.append(end_time-start_time)
```

```python
final = pd.DataFrame(list(zip(candidates, results, times)))
final.columns = ['Candidates', 'Generations', 'Time']

plt.title('Numbers of candidates compared to generations to solve')
plt.plot(final['Candidates'], final['Generations'])
plt.grid()
plt.ylabel('Generations')
plt.xlabel('Candidates')
plt.show()
```

```python
plt.title('Numbers of candidates compared to time to solve')
plt.plot(final['Candidates'], final['Time'])
plt.grid()
plt.ylabel('Time')
plt.xlabel('Candidates')
plt.show()
```

## Mutation rate

```python
mutation_rates = [0.02, 0.05, 0.1, 0.15, 0.2, 0.4, 0.5, 0.7, 0.8, 1]
results = list()
times = list()

for mut in mutation_rates:
    temp = list()
    start_time = time.time()
    for _ in range(10):
        temp.append(solve(mutation_rate=mut))
    results.append(np.mean(temp))
    end_time = time.time()
    times.append(end_time-start_time)
```

```python
final = pd.DataFrame(list(zip(mutation_rates, results, times)))
final.columns = ['Mutation rate', 'Generations', 'Time']

plt.title('Mutation rate compared to generations to solve')
plt.plot(final['Mutation rate'], final['Generations'])
plt.grid()
plt.ylabel('Generations')
plt.xlabel('Mutation rate')
plt.show()
```

```python
plt.title('Mutation rate compared to time to solve')
plt.plot(final['Mutation rate'], final['Time'])
plt.grid()
plt.ylabel('Time')
plt.xlabel('Mutation rate')
plt.show()
```
